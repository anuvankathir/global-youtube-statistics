```python
import pandas as pd

# Load your data into a DataFrame
df = pd.read_csv('dataset.csv', encoding='utf-8')

# Iterate over each column
for column in df.columns:
    # Check if the column contains string data type
    if df[column].dtype == 'object':
        # Clean non-ASCII values from string columns
        df[column] = df[column].apply(lambda x: ''.join([i if ord(i) < 128 else '' for i in str(x)]))

# Now your DataFrame should be cleaned, with only ASCII values in string columns
```
